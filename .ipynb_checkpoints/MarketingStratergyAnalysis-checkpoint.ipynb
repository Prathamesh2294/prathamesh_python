### Importing libraries

```python
import pandas as pd
import os 
import matplotlib.pyplot as plt
```

#### Merge three months data to form single DataFrame 

```python
new_df=pd.DataFrame()
files = [file for file in os.listdir("C:\\Program Files\\Python36\\suven\\Notebooks")]
for file in files:
    if(file.endswith("_sales.csv")):
        df=pd.read_csv("C:\\Program Files\\Python36\\suven\\Notebooks\\"+file)
        new_df=pd.concat([new_df,df])
new_df.to_csv("complete_data.csv",index=False)
new_df.head()
```

### Data Cleaning Process

##### Task 1: Number of NaN Columnwise 

```python
new_df.isnull().sum(axis=0)
```

##### Task 2: Dropping  the Columns

```python
new_df=new_df.drop(['id','enquiry_id','data','editage_card_id','invoice_id','created_by','last_modified_by','last_modified_date','payment_admin_eos_id','component_id','converted_amount','converted_currency'],axis=1)
new_df.head()
```

#### Task 4: Adding Month Column using Data Slicing Concept

```python
new_df['month']=new_df['created_date'].str[5:7]
new_df['month']=new_df['month'].astype('int32')
new_df.to_csv("complete_data.csv",index=False)
new_df.head()
```

#### Problem Statement 1: What was the Best month for Sales?How Much do we earn in the particular month? 

```python
result=new_df.groupby('month').sum()['price_after_tax']
print(result)
```

#### Problem Statement 2: Which partner had given highest number of business 

```python
result['price_after_tax']=new_df.groupby('currency_code').sum()['price_after_tax']
result['price_after_tax'] = result['price_after_tax'].apply(lambda x: '{:.2f}'.format(x))
temp_result=result['price_after_tax']
print(temp_result)
```

#### Problem Statement 3:What time should we display advertisements to increase likehood for customer sending documents?

```python
new_df['created_date']=pd.to_datetime(new_df['created_date'])
new_df['hour']=new_df['created_date'].dt.hour
new_df['minute']=new_df['created_date'].dt.minute
new_df.head()
hours =[hour for hour,df in new_df.groupby('hour')]
plt.plot(hours,new_df.groupby(['hour']).count())
plt.show()
```
